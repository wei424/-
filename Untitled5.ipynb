```python
import matplotlib.pyplot as plt
import numpy as np

# 定義函數
def f(x):
    return -2*x*x + 4*x + 5

# 生成 x 值
x = np.linspace(1, 100, 100)

# 計算對應的 y 值
y = f(x)

# 繪製圖表
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = -2*x*x + 4*x + 5')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 設定函式參數
a = 3
b = 2
c = -6

# 生成 x 值範圍
x = np.linspace(1, 100, 100)

# 計算對應的 y 值
y = a * x**2 + b * x + c

# 繪製函式圖形
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = {}*x^2 + {}*x + {}'.format(a, b, c))
plt.grid(True)
plt.show()
```
